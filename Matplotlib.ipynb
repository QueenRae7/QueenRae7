# It is a Data Visualisation library used to plot some basics chart so that we can quickly analyse our dataset

```python
# Loading the library
import matplotlib.pyplot as plt
```

## 1. Line Chart --- it will show the linear trend of data

```python
Min_Temp = [30, 50, 34, 47, 32, 45]
Days = [1, 2, 3, 4, 5, 6]

plt.plot(Days, Min_Temp)
plt.show()
```

```python
plt.figure(figsize= (8, 5))
Min_Temp = [30, 50, 34, 47, 32, 45]
Days = [1, 2, 3, 4, 5, 6]

plt.plot(Days, Min_Temp, color = 'red')
plt.xlabel("Number of Days")
plt.ylabel("Temperature")
plt.title("Days vs Temperature")
plt.show()
```

```python
plt.figure(figsize= (8, 5))
Min_Temp = [30, 50, 34, 47, 32, 45]
Days = [1, 2, 3, 4, 5, 6]

plt.plot(Days, Min_Temp, color = 'red', marker = "*", linestyle = "-.")
plt.xlabel("Number of Days")
plt.ylabel("Temperature")
plt.title("Days vs Temperature")
plt.show()
```

```python
# Multiple chart in same figure
plt.figure(figsize= (8, 5))
Min_Temp = [30, 50, 34, 47, 32, 45]
Max_Temp = [40, 52, 39, 51, 34, 47]
Days = [1, 2, 3, 4, 5, 6]

plt.plot(Days, Min_Temp, color = 'red', marker = "*", linestyle = "-.")
plt.plot(Days, Max_Temp, color = 'green', marker = "D", linestyle = ":")
plt.xlabel("Number of Days")
plt.ylabel("Temperature")
plt.title("Days vs Temperature")

plt.legend(["Min_Temp", "Max_Temp"])
plt.show()
```

```python
# Adding the grid

# Multiple chart in same figure
plt.figure(figsize= (8, 5))
Min_Temp = [30, 50, 34, 47, 32, 45]
Max_Temp = [40, 52, 39, 51, 34, 47]
Days = [1, 2, 3, 4, 5, 6]

plt.plot(Days, Min_Temp, color = 'red', marker = "*", linestyle = "-.")
plt.plot(Days, Max_Temp, color = 'green', marker = "D", linestyle = ":")
plt.xlabel("Number of Days")
plt.ylabel("Temperature")
plt.title("Days vs Temperature")

plt.legend(["Min_Temp", "Max_Temp"])
plt.grid(alpha = 0.4, color = 'pink')
plt.show()
```
