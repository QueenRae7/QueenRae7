## iloc and loc selectors

They are used for making the slicing in the pandas dataframe.
.loc follow the index whereas the loc follows the label

```python
Dataset1.head(10)
```

```python
# iloc
Dataset1.iloc[2:6, 2:6]
```

```python
#loc
Dataset1.loc[2:5, "sex":"time"]
```

```python
Dataset1.head(10)
```

```python
# iloc
Dataset1.iloc[2:6, 0:7]
```

```python
Dataset1.iloc[2:6, :]
```

```python
# loc
Dataset1.loc[2:5, "total_bill":"size"]
```
