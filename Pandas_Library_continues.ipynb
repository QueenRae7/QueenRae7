## 2. Appraoch - Filling the missing values

```python
Dataset1.isnull().sum()
```

```python
Dataset1.dtypes
```

### Note- if the dtype of column is Numerical then filled with Median/mean() and if the dtype of the column is text then fill with mode()

```python
Dataset1.columns
```

```python
for col in Dataset1.columns:
  if Dataset1[col].dtypes == 'object':
    Dataset1[col] = Dataset1[col].fillna(Dataset1[col].mode()[0])
  else:
    Dataset1[col] = Dataset1[col].fillna(Dataset1[col].median())
```

```python
Dataset1.isnull().sum()
```

# Dealing with Duplicate Records

```python
My_Data1 = pd.read_csv("/content/diabetes_D.csv")
My_Data1.head()
```

```python
# Checking the duplicate records in dataset
My_Data1.duplicated().sum()
```

```python
My_Data1.shape
```

```python
# Displaying the row having the duplicate records
My_Data1.loc[My_Data1.duplicated(), :]
```
